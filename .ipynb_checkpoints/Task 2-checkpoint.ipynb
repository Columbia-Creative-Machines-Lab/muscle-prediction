```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_covtype
```

## Import and store data

```python
data = fetch_covtype()
df = pd.DataFrame(data['data'])
df = df.join(pd.DataFrame(data['target'], columns=['Target']))
#df = df.head(n=10000) # For testing purposes only
df = df.sample(n=10000) # Some machines are not able to run this
                        # notebook over the entire 580k+ row dataset, 
                        # so we might randomly sample to improve runtime.
```

## Q2.1: Visualize feature and target univariate distributions

```python
fig, axes = plt.subplots(ncols=7, nrows=8, figsize=(30,30))
for i in range(0, 55):
    try:
        sns.distplot(df.iloc[:, i], ax=axes.flat[i])
    except:
        pass # Ignore LinAlgError for singular matrices
axes.flat[-1].set_visible(False)
```

## Q2.3: Split data and evaluate models
_We observe that the results of logistic regression, linear SVMs, and nearest centroids are very different. Logistic regression performs the best with default cross-validation parameters._

```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.neighbors import NearestCentroid
```

```python
X_train, X_test, y_train, y_test =\
    train_test_split(df.iloc[:, :-1], df.Target)
```

### Evaluate models using cross-validation with default parameters

```python
clfs = [LogisticRegression(solver='lbfgs', multi_class='multinomial'),\
        LinearSVC(dual=False),\
        NearestCentroid()]
scores = [np.mean(cross_val_score(clf, X_train, y_train)) for clf in clfs]
scores
```

### Scale data and reevaluate models
_We observe that scaling the data with StandardScaler does improve model accuracy for all models, especially linear SVM and nearest centroid._

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
scaled_scores = [np.mean(cross_val_score(clf, X_train_scaled, y_train))\
                 for clf in clfs]
scaled_scores
```

## Q2.3: Tune model parameters with GridSearchCV
_We observe that results do improve with model parameters tuned by_ `GridSearchCV`.

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': np.linspace(np.float_power(10, -3),\
                               np.float_power(10, 3), num=10)}
nc_param_grid = {'shrink_threshold': np.linspace(-1.0, 1.0, num=10)}

grids = [GridSearchCV(clf, param_grid=param_grid, return_train_score=True)\
         for clf in clfs[:2]]
grids.append(GridSearchCV(clfs[-1], param_grid=nc_param_grid,\
                         return_train_score=True))
grids = map(lambda grid: grid.fit(X_train, y_train), grids);
```

```python
best_scores = [grid.best_score_ for grid in grids]
best_scores
```

```python
best_params = [grid.best_params_ for grid in grids]
best_params
```

## Visualize dependence of validation score on model parameters

```python
# Clean data from the cv_results_ attribute of each GridSearchCV object
def get_score_and_params(raw_results, param):
    params = pd.DataFrame([d[param] for d in raw_results['params']])
    params.columns = [param]
    scores = pd.DataFrame(raw_results['mean_train_score'])
    scores.columns = ['mean_train_score']
    return params.join(scores)
```

```python
clf_names = ['Logistic Regression', 'Linear SVM', 'Nearest Centroid']
fig, axes = plt.subplots(ncols=3, nrows=1, figsize=(18,4))

sns.regplot('C', 'mean_train_score', get_score_and_params(grids[0].cv_results_, 'C'),\
            fit_reg=False, ax=axes.flat[0],\
            scatter_kws={'s':25}).set_title(clf_names[0])
sns.regplot('C', 'mean_train_score', get_score_and_params(grids[1].cv_results_, 'C'),\
            fit_reg=False, ax=axes.flat[1],\
            scatter_kws={'s':25}).set_title(clf_names[1])
plt.xticks(np.linspace(-1.0, 1.0, num=10), rotation=45)
plt.xlim(-1.1, 1.1)
sns.regplot('shrink_threshold', 'mean_train_score',\
            get_score_and_params(grids[2].cv_results_, 'shrink_threshold'),\
            fit_reg=False, ax=axes.flat[2],\
            scatter_kws={'s':25}).set_title(clf_names[2]);
```

### Q2.4: Change cross-validation strategy from `stratified k-fold` to `kfold` with shuffling
_We observe that the parameters found by k-fold cross-validation have changed, compared to those found by stratified k-fold._

```python
from sklearn.model_selection import KFold
```

```python
k_fold = KFold(shuffle=True)
kf_grids = [GridSearchCV(clf, param_grid=param_grid, cv=k_fold) for clf in clfs[:2]]
kf_grids.append(GridSearchCV(clfs[-1], param_grid=nc_param_grid, cv=k_fold))
kf_grids = map(lambda grid: grid.fit(X_train, y_train), kf_grids);
```

```python
best_scores = [grid.best_score_ for grid in kf_grids]
best_scores
```

```python
best_params = [grid.best_params_ for grid in kf_grids]
best_params
```

### Evaluate model performance with new shuffling random seed
_We osberve that parameters found after adjusting the random seed of the k-fold shuffle have changed._

```python
k_fold = KFold(shuffle=True, random_state=3)
grids = [GridSearchCV(clf, param_grid=param_grid, cv=k_fold) for clf in clfs[:2]]
grids.append(GridSearchCV(clfs[-1], param_grid=nc_param_grid, cv=k_fold))
seed_grids = map(lambda grid: grid.fit(X_train, y_train), grids);
```

```python
best_scores = [grid.best_score_ for grid in seed_grids]
best_scores
```

```python
best_params = [grid.best_params_ for grid in seed_grids]
best_params
```

### Evaluate model performance with modified training/test split
_We observe that some of the parameters found after adjusting the training/test split to 80/20 have changed._

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, :-1],\
                                                    df.Target,\
                                                    train_size=0.8,\
                                                    test_size=0.2)
split_grids = map(lambda grid: grid.fit(X_train, y_train), grids);
```

```python
best_scores = [grid.best_score_ for grid in split_grids]
best_scores
```

```python
best_params = [grid.best_params_ for grid in split_grids]
best_params
```

## Q2.5: Visualize model coefficients

```python
fig, axes = plt.subplots(ncols=4, nrows=2, figsize=(30,15))

for i in range(0, 2):
    coefs = grids[i].best_estimator_.coef_
    for j in range(0, coefs.shape[0]):
        sns.regplot(np.arange(0, coefs.shape[1]), coefs[j],\
                    ax=axes.flat[j], fit_reg=False,\
                    label=clf_names[i], scatter_kws={'alpha': 0.8})
        axes.flat[j].legend();
axes.flat[-1].set_visible(False)
```
